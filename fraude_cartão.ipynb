```python
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler
from imblearn.pipeline import Pipeline
```

```python
#Para resolver o problema de desbalanceamento será utilizado o undersampling and oversampling, o over será de 10%, ou seja a classe minonitária(1) terá 10% do tamanho da
#majoritaria(0). O under será de 30%, ou seja a classe majoritaria(0) terá 30% a mais de tamanho da classe minonitária(1)
over = SMOTE(sampling_strategy=0.1)
under = RandomUnderSampler(sampling_strategy=0.3)
steps = [('o', over), ('u', under)]
pipeline = Pipeline(steps=steps)
X1, y1 = pipeline.fit_resample(X, y)
```
